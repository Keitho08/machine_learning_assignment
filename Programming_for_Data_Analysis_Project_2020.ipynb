## 4.1 Creating Synthesised Data

In creating our synthesised data we first import any modules required below.

```python
import numpy as np # imports numpy as np
import pandas as pd # imports pandas as pd
import datetime # The datetime module supplies classes for manipulating dates and times.
import math # used for carrying out any python math functions we may need
import matplotlib.pyplot as plt # imports matplotlib.pyplot as plt
import matplotlib.dates as mdates # matplotlib.dates is used for plotting dates data
import seaborn as sns # seaborn imported as sns
```

Our next step is to create our date datasets, this can be achieved using the code below using the datetime.datetime function and also pandas date_range function. A data set of 365 days is created for each year 2018-2020.

*Note: As 2020 is a leap year we made the end date 30th December so that we could get same size arrays for each year. This was also noted initially when errors occured due to differences in size arrays. Following some troubleshooting it was found to be the leap year causing issues!*

```python
# Creating the Date's for 2018:
start2018 = datetime.datetime(2018, 1, 1) # start date format (year, month, day)
end2018 = datetime.datetime(2018, 12, 31) # end date
dates2018 = pd.date_range(start2018, end2018, freq='d') # date range in frequency of days. 

# Creating the Date's for 2019:
start2019 = datetime.datetime(2019, 1, 1)
end2019 = datetime.datetime(2019, 12, 31)
dates2019 = pd.date_range(start2019, end2019, freq='d')

# Creating the Date's for 2020:
start2020 = datetime.datetime(2020, 1, 1)
end2020 = datetime.datetime(2020, 12, 30) # leap year so we ignore last day to keep all arrays same size
dates2020 = pd.date_range(start2020, end2020, freq='d')
```

Having created the date range above we are now ready to create the Covid Cases dataset. We can use np.random.normal for 2018 and 2019, with a mean of 0 (loc) and standard deviation of 0 (scale) to create 365 0 values. 2020 was a small bit trickier and required some experimentation with the numpy random package. After some trial and error it was found the best way to create the 2020 dataset was to use the random.randint function setting a max value of 450 and size to 365 allowed us to create an array with 365 values between 0 and 450. Using the seed of 1, this data set allowed us to achieve a total covid figure quite close to official goverment figure as of 17th December 2020 which was 77,678 (https://www.gov.ie/en/news/7e0924-latest-updates-on-covid-19-coronavirus/). Our total figure simulated for 2020 was 80,227 which is a very close match considering there are 14 days left in 2020 at the time of writing. 

```python
# Creating the Covid Cases
np.random.seed(1) # sets the seed to 1 to generate same results each run
synCovidDailyCases2020 = np.random.randint(450, size=365) # generates an array of 365 random integers from 0 to 450
synCovidDailyCases2018 = np.random.normal(loc=0, scale=0, size=365) # generates an array of 365 0 values
synCovidDailyCases2019 = np.random.normal(loc=0, scale=0, size=365) # generates an array of 365 0 values
print(synCovidDailyCases2020) # prints out the array of covid cases
synTotal2020CovidCases = np.sum(synCovidDailyCases2020) # sums up all values in array above
print('Total number of Covid Cases in 2020 is', synTotal2020CovidCases) # prints out total number of cases in 2020
```

Creating the dataset for the weather data was challenging. On initial attempts various iterations of normal distributions were created, however the output results did not appear to be realistic and there was a considerable difference from actual data recorded for 2020. However, following many hours online research, discussions with users on Stack Exchange and user experimentation it was found that the best algorithim to produce the weather data with similar output to reference weather was:

$ synMintp2020 = np.sin(np.arange(365)/365 * math.pi)*trange + tmin $

In the above equation math.pi is equal to the pi value abbreviated as 3.14159265359 and trange is equal to the standard deviation of the temperature recored in 2020 and tmin is the minimum temperature recorded in 2020. It was also experimented to use the tmean (average temperature in 2020) instead of the tmin, however the tmin produced a graph that was much closer to original data set.

![alt text.](https://lh3.googleusercontent.com/oTFhGvMxu3eXpFJAB2Omb8CHRTBI_JEL7sZ6qIc8GNBa0mYUjdGCiF4yVKtyyHhveAA3k8jx5_HQNQPqXocg5rX9cshsdm9AIIT2VuCzZADk9rO79XetUK7-VxbeqhA=s412 "Sine Weather")
**Image Source:**  https://lh3.googleusercontent.com/oTFhGvMxu3eXpFJAB2Omb8CHRTBI_JEL7sZ6qIc8GNBa0mYUjdGCiF4yVKtyyHhveAA3k8jx5_HQNQPqXocg5rX9cshsdm9AIIT2VuCzZADk9rO79XetUK7-VxbeqhA=s412

```python
# Creating the Weather Data
np.random.seed(1) # sets the seed to generate similar results each iteration
t_range = 4.383604 # our range (standard deviation)
t_min = -5.0 # our minimum value

synMintp2020 = np.sin(np.arange(365)/365 * math.pi)*t_range + t_min # applies the formula described above
synMintp2020 += np.random.normal(loc=7.203288, scale=3, size=365) 
# generates a normal distribution, loc=range of data set, scale is set to 3 as this generates a good fit of the data
# based on experimentation. += adds a number from the normal distribution to the synMintp2020 variable

syn2020MintpData = pd.DataFrame({'Date': dates2020, 'MinTp': synMintp2020}) # creates a dictionary with weather data
sns.scatterplot(x="Date", y='MinTp', data=syn2020MintpData[0:], color='g')
# creates a scatterplot from dictionary
plt.title('Simulated Minimum Outdoor Air Temperature - 2020', fontsize=20)
```

For the last section of code for this section we will create our random energy data for each year. For years 2018 and 2019 we use the mean value (loc) and standard deviation (scale) from the real life data sets to generate random normal distributions for each of the two years. 

For the 2020 data set we wanted to simulate the effect of the weather and Covid cases on the data set. The following algorithim was developed to achieve this:

$ synEne2020 = synEne2019 x  1.0 + (synCovidDailyCases2020 x 0.03) - synMintp2020 $

In this equation we assumed that 2020 would be comparable to 2019 prior to adding the Covid cases where we assumed that we would add 3% of the total number of cases each day to annual energy usage in GWh (Giga Watts Hours). To simulate the effect of the weather we subtracted the outdoor air temperature from the total energy for each day. The logic to this was that we assumed the higher the outdoor temperature the less usage would be required on artifical lighting as it would likely relate to increased outdoor daylight levels. 

```python
# Creating the Energy for 2018:
np.random.seed(3) # This particular seed generates a good spread of the data
synEne2018 = np.random.normal(loc=66.883795, scale=5.448145, size=365) # creates normal dist. for Energy 2018
syn2018EnergyDict = pd.DataFrame({'Date': dates2018, 'TotalDailyEnergy_GWh': synEne2018}) # Creates a dictionary for energy and dates 2018
# Creating the Energy for 2019:
synEne2019 = np.random.normal(loc=66.090044, scale=5.620020, size=365)
syn2019EnergyDict = pd.DataFrame({'Date': dates2019, 'TotalDailyEnergy_GWh': synEne2019})
# Creating the Energy for 2020:
synEne2020 = synEne2019 *  1.0 + (synCovidDailyCases2020 * 0.03) - synMintp2020
# synEne2020 = synEne2019 *  1.05 + (synCovidDailyCases2020 * 0.01)
syn2020EnergyDict = pd.DataFrame({'Date': dates2020, 'TotalDailyEnergy_GWh': synEne2020}) 
# Creating a combined Energy Set
synCombinedData = pd.concat([syn2018EnergyDict, syn2019EnergyDict, syn2020EnergyDict]) # combines all above 3 dictionaries into one data set.
```

```python
# Creates our dictionary for the data required
EnergyvsCovid = pd.DataFrame({'Date': dates2020, 'ConfirmedCovidCases': synCovidDailyCases2020, 'Daily Energy': synEne2020})
EnergyvsCovid['datetime'] = pd.to_datetime(EnergyvsCovid['Date']) # converts the date column into datetime format for pandas to read as a date field
EnergyvsCovid

#Create combined plot parameters
fig, ax1 = plt.subplots(figsize=(18,10))
#line plot creation 1
ax1.set_title('Covid Cases 2020, Energy Usage 2020', fontsize=16)
ax1.set_xlabel('Date', fontsize=16)
ax1.set_ylabel('Daily Energy ', fontsize=16)
ax1 = sns.kdeplot(x='datetime', y='Daily Energy', data=EnergyvsCovid, color='g', thresh=.1, label='Dailyenergy') 
# thresh = Lowest iso-proportion level at which to draw a contour line. 
#specify we want to share the same x-axis
ax2 = ax1.twinx()
#line plot creation 2
ax2 = sns.kdeplot(x='datetime', y='ConfirmedCovidCases', data=EnergyvsCovid, color='r', thresh=.1, label='CovidCases')
#show plot
plt.legend()
plt.show()
```

We can see from the output kde plot that the peaks for Covid cases and daily energy usage are closely related indicating to us that Covid 19 cases have a direct impact on our energy usage. Let's look at the individual points from above kde plot by using a scatter plot below:

```python
# Creates our dictionary for the plot
EnergyvsCovid = pd.DataFrame({'Date': dates2020, 'ConfirmedCovidCases': synCovidDailyCases2020, 'Daily Energy': synEne2020})
EnergyvsCovid['datetime'] = pd.to_datetime(EnergyvsCovid['Date']) # converts the date column into datetime format for pandas to read as a date field
EnergyvsCovid

#Create combined plot parameters
fig, ax1 = plt.subplots(figsize=(18,10))
#line plot creation 1
ax1.set_title('Covid Cases 2020, Energy Usage 2020', fontsize=16)
ax1.set_xlabel('Date', fontsize=16)
ax1.set_ylabel('Daily Energy ', fontsize=16)
ax1 = sns.scatterplot(x='datetime', y='Daily Energy', data=EnergyvsCovid, color='g', label='DailyEnergy') 
#specify we want to share the same x-axis
ax2 = ax1.twinx()
#line plot creation 2
ax2 = sns.scatterplot(x='datetime', y='ConfirmedCovidCases', data=EnergyvsCovid, color='r', marker='x', label='CovidCases')
#show plot
plt.legend()
plt.show()
```

Following on from our previous plot, the above scatter plot shows the individual points for Covid cases vs. energy usage 2020. We will now examine the linear relationship between the two variables using a seaborn lmplot. A seaborn lmplot combines a regression plot with a facetgrid for the data set. It is a convenient build in method to seaborn library to fit a regression model on this dataset. The code required is as below.

```python
# Create our dictionary for the data and add datetime field:
EnergyvsCovid = pd.DataFrame({'Date': dates2020, 'ConfirmedCovidCases': synCovidDailyCases2020, 'Daily Energy': synEne2020})
EnergyvsCovid['datetime'] = pd.to_datetime(EnergyvsCovid['Date']) # converts the date column into datetime format for pandas to read as a date field
EnergyvsCovid

sns.lmplot(x="ConfirmedCovidCases", y="Daily Energy", data=EnergyvsCovid, height=8, aspect=2)
# creates an lmplot with above x, y and data inputs and sets the ratio for height and aspect to set size of plot
```

We can see from our lmplot that there is a good positive relationship between increase in daily Covid cases and daily energy usage, confirming our previous hypotheses. This is leading to a higher number of maximum energy usage than the numbers recorded in previous years. 

**Energy vs Weather**

We can see from above line plots that the daily energy usage seems to reduce as the daily weather increases. However, lets look at this relationship more linearly using another lmplot below:

```python
WeatherEnergyRelationship=sns.lmplot(x="Weather", y="Daily Energy", data=EnergyvsWeather, height=8, aspect=2)
plt.show()
```

From our lmplot we can see there is a negative relationship between energy usage and increased outdoor air temperature. This would lead us to believe that higher outdoor temperatures, which are seasonal and occur during the summer months (as per previous plots in this notebook), also relate to increased daylight hours. The reduced energy is likely due to a combination of reduction in use of artifical lighting and also reduction in energy required for heating, ventilation and air conditioning plant.
